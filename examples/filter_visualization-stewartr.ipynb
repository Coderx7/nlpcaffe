```python
from __future__ import division
import numpy as np
import pickle
from tsne import bh_sne
%matplotlib inline
import matplotlib.pyplot as plt
import math
import lmdb
import itertools


import mpld3
mpld3.enable_notebook()
from mpld3 import plugins


# Make sure that caffe is on the python path:
caffe_root = '/home/stewartr/git/caffe_house/caffe3/'  # this file is expected to be in {caffe_root}/examples
import sys
sys.path.insert(0, caffe_root)
sys.path.insert(0, caffe_root + 'python')
sys.path.append(caffe_root + '/models/rnn')
import caffe_pb2
import caffe
import mkproto

#plt.rcParams['figure.figsize'] = (10, 10)
#plt.rcParams['image.interpolation'] = 'nearest'
#plt.rcParams['image.cmap'] = 'gray'
```

```python
net = caffe.Classifier(caffe_root + 'models/rnn/deploy.prototxt',
                       '/snapshots/caffe_iter_80000.caffemodel')
net.set_phase_test()
net.set_mode_gpu()
```

```python
vocab = pickle.load(open('/home/stewartr/data/zhen/vocab.pkl', 'r'))
inv = lambda d: {v:k for k,v in d.iteritems()}
vocab_inv = inv(vocab)

def get_datums():
    env = lmdb.open(caffe_root + '/models/rnn/rnn_test_db')
    with env.begin() as txn:
        cursor = txn.cursor()
        while cursor.next():
            datum = caffe_pb2.Datum()
            datum.ParseFromString(cursor.value())
            yield datum

#datums = list(itertools.islice(get_datums(), 3000))
datums = list(get_datums())
```

```python
def get_data(datums):
    n = datums[0].channels // 2
    for i in range(0, len(datums), mkproto.deploy_batch_size):
        data = np.zeros((mkproto.deploy_batch_size, 2 * n, 1, 1))
        for j in range(mkproto.deploy_batch_size):
            if (i + j) >= len(datums):
                return
            data[j, :, 0, 0] = datums[i + j].float_data
        yield data

data_batches = list(get_data(datums))
```

```python
def score_sentence(datum, net, net_id):
    global_loss = np.zeros(mkproto.target_length,)
    local_loss = np.zeros(mkproto.target_length,)
    lengths = np.zeros(mkproto.target_length,)
    for i in range(mkproto.target_length):
        target = int(datum.float_data[mkproto.source_length + mkproto.target_length + i] + 0.5)
        target_category_id = int(net.blobs['target_category_id'].data[net_id + i * mkproto.deploy_batch_size].flatten()[0])
        target_local_id = int(net.blobs['target_local_id'].data[net_id + i * mkproto.deploy_batch_size].flatten()[0])
        assert target_category_id == target % mkproto.num_categories
        assert target_local_id == target // mkproto.num_categories
        global_loss[i] = -math.log(net.blobs['category_prob'].data[net_id + mkproto.deploy_batch_size * i, target_category_id, 0, 0])
        local_loss[i] = -math.log(net.blobs['local_prob'].data[net_id + mkproto.deploy_batch_size * i, target_local_id, 0, 0])

        lengths[i] += 1
        if target == mkproto.t_zero_symbol:
            break
    return local_loss, global_loss, lengths
```

```python
global_losses = []
local_losses = []
lengths = []
for i, batch in enumerate(data_batches):
    net.rff(batch)
    for j in range(len(batch)):
        datum = datums[j + mkproto.deploy_batch_size * i] 
        local_loss, global_loss, length = score_sentence(datum, net, j)
        global_losses.append(global_loss)
        local_losses.append(local_loss)
        
        lengths.append(length)
        #print (global_loss , local_loss) #/ length
    print '%d' % i
    print 'local: %s' % (np.sum(local_losses, 0) / np.sum(lengths,0))
    print 'global: %s' % (np.sum(global_losses, 0) / np.sum(lengths,0))
    
```

```python
word_vectors = {k:net.params['wordvec_layer'][0].data[0, vocab[k], 0, :] for (k, v) in vocab.items() if v < mkproto.vocab_size}
M = np.zeros((len(word_vectors), len(word_vectors.values()[0].flatten())))
for k, v in word_vectors.items():
    M[vocab[k], :] = v
    
X_2d = bh_sne(M)
```

```python
fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'), figsize=(8,8))
ax.grid(color='white', linestyle='solid')

num_words_to_plot = 100

scatter = ax.scatter(X_2d[:num_words_to_plot, 0], X_2d[:num_words_to_plot, 1], s = 100, c='#3EEEEE', alpha=0.1, cmap=plt.cm.jet)
labels = [vocab_inv[i] for i in range(num_words_to_plot)]
plugins.connect(fig, plugins.PointLabelTooltip(scatter, labels))
```
